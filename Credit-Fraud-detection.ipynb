```python
# importing the dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# load the credit data
credit_data = pd.read_csv('/content/creditcard.csv')
```

```python
# Checking the missing values
credit_data.isnull().sum()
```

```python
# counting the distribution of legit and fraud transaction\
credit_data['Class'].value_counts()

# very unbalaced dataset-- we cant feed this data into machine learing model,
# because very less data-point fro the fraud
# -----have to process the dataset further
```

```python
# separating the dataset
legit =  credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# comparing the values for both transactions
credit_data.groupby('Class').mean()
```

```python
# under-sampling of the data
# build a sample dataset containing similar distribution
legit_sample = legit.sample(n=492)
```

```python
# concat 2 data-frame (legit_sample and fraud)
new_data = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
# splitting the data into input and targets
X = new_data.drop(columns='Class', axis=1)
Y = new_data['Class']
```

```python
print(Y)
```

```python
print(X)
```

```python
# splitting the data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Training the model with the logistic regression model
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
# Evaluating the model based oon the accuracy score
X_train_pred = model.predict(X_train)
data_accuracy = accuracy_score(X_train_pred, Y_train)
```

```python
data_accuracy
```

```python
# Accuracy score for the testing data
X_test_pred = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_pred, Y_test)
```
